```python
import pandas as pd
```

```python
flag_data = "static/data/country_flags.csv"
country_data = "static/data/country_data.csv"

#read files as dataframe

flag_df = pd.read_csv(flag_data)

country_df = pd.read_csv(country_data)

country_df.head()
```

```python
flag_df.head()
```

```python
#remove columns from country_df that are not of interest

country_df = country_df.drop(columns=["Service","Industry","Agriculture","Deathrate","Birthrate","Climate","Other (%)","Crops (%)",
                                      "Arable (%)","Phones (per 1000)","Infant mortality (per 1000 births)","Literacy (%)",
                                      "Net migration","Coastline (coast/area ratio)","Pop. Density (per sq. mi.)"])

country_df.head()
```

```python
#merge the two dataframes on country name

combo_df = pd.merge(country_df, flag_df[["Country","ImageURL"]], on="Country")

combo_df.head()
```

```python
#clean up the Region column - create broader continent groups instead of regional groups


combo_df["Region"].replace(to_replace="EASTERN EUROPE", value="Europe", inplace=True)
combo_df["Region"].replace(to_replace="WESTERN EUROPE", value="Europe", inplace=True)
combo_df["Region"].replace(to_replace="ASIA (EX. NEAR EAST)", value="Asia", inplace=True)
combo_df["Region"].replace(to_replace="NORTHERN AFRICA", value="Africa", inplace=True)
combo_df["Region"].replace(to_replace="SUB-SAHARAN AFRICA", value="Africa", inplace=True)
combo_df["Region"].replace(to_replace="LATIN AMER. & CARIB", value="North America", inplace=True)
combo_df["Region"].replace(to_replace="OCEANIA", value="Oceania", inplace=True)
combo_df["Region"].replace(to_replace="NORTHERN AMERICA", value="North America", inplace=True)
combo_df["Region"].replace(to_replace="NEAR EAST", value="Asia", inplace=True)
combo_df["Region"].replace(to_replace="BALTICS", value="Europe", inplace=True)

combo_df.head(10)
```

```python
#Weird category Commonwealth of Independent States which has overlap between Europe and Asia.

CIS = combo_df.loc[combo_df['Region'] == "C.W. OF IND. STATES"]

CIS
```

```python
#Replace the regions for the CIS

combo_df.loc[[6,9,56,79,83,154,161,169], 'Region'] = 'Asia'

combo_df.loc[[13,104,130,164], 'Region'] = 'Europe'
```
