```python
import json
import pandas as pd
import requests
import pprint
```

```python
#get json file from website, download & save

url = requests.get('https://flagpedia.net/download/country-codes-case-lower.json')
data = url.json()

with open('static/json/countries.json', 'w') as f:
    json.dump(data, f)
```

```python
code_df = pd.read_json('static/json/countries.json', orient='index')

code_df.head()
```

```python
code_df = code_df.reset_index()

code_df = code_df.rename(columns={0:"Country", "index":"Code"})

code_df.head()
```

```python
url_list = []

for i in code_df["Code"]:
    url = "static/images/"+i+".svg"
    url_list.append(url)
    
url_list
```

```python
code_df["URL"] = url_list

code_df.tail()
```

```python
code_df.to_csv("static/data/country_codes.csv")
```
